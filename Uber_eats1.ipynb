1. Before analyzing the experiment’s results, we want to verify that the experiment
randomized users properly. To do this, we need to compare the treatment vs. control
groups on their users’ baseline characteristics — this is often referred to as a balance
test or a randomization check. (5 points)

a. Compare the proportion of American users (more specifically: users living in
America) in the treatment vs. control groups by using a t-test. Interpret the
result and its statistical significance. What do you conclude about the validity
of the experimental randomization in terms of country of residence?

b. Compare the proportion of American users (more specifically: users living in
America) in the treatment vs. control groups for users who saw ads by using
a t-test. Interpret the result and its statistical significance. What do you
conclude from this?

c. Compare the average past revenue in the treatment vs. control groups by
using a t-test. Interpret the result and its statistical significance. What do you
conclude about the validity of the experimental randomization in terms of
past revenue?

d. Compare the average past revenue in the treatment vs. control groups for
users who saw ads by using a t-test. Interpret the result and its statistical
significance. What do you conclude from this?

e. Apart from a t-test, what additional analysis could be used to verify that Uber
Eats correctly randomized users into the treatment vs. control groups with
regards to their average past revenue?

```python
#1.a
treatment = df[df['treatment'] ==1]
control = df[df['treatment'] ==0]
```

```python
#1.a
from scipy.stats import ttest_ind
treatment['is_american'] = (treatment['country'] == 'american').astype(int)
control['is_american'] = (control['country'] == 'american').astype(int)
```

```python
#1.a
ttest_ind(treatment['is_american'], control['is_american'])
# p value is 1.0, we fail to reject null hypotsesis,
#that means randomization is successfull and valid.
```

```python
#1.b
treatment['american+ad'] = ((treatment['country'] == 'american') & (treatment['saw_ads'] == 1)).astype(int)
control['american+ad'] = ((control ['country'] == 'american') & (control['saw_ads'] == 1)).astype(int)
```

```python
#1.b
ttest_ind(treatment['american+ad'], control['american+ad'])
# Since the p-value (1.0) is much greater than the
# significance level of 0.05
#we fail to reject the null hypothesis.
# This indicates that the randomization was successful
# in terms of balancing American users who
#saw ads across both the treatment and control groups.
```

```python
#1.c
treatment['revenue'] = treatment['past_revenue']
control['revenue'] = control['past_revenue']
```

```python
#1.c
ttest_ind(treatment['revenue'],control['revenue'])
# pvalue is higher than 0.05 so randomization check passed.
```

```python
#1.d
treatment_Ads = treatment[treatment['saw_ads'] == 1]
control_Ads = control[control['saw_ads'] == 1]
treatment_Ad_rev = treatment_Ads['past_revenue']
control_Ad_rev = control_Ads['past_revenue']
```

```python
#1.d
ttest_ind(treatment_Ad_rev, control_Ad_rev)
#pvalue is 0.7 which is higher than 0/05, so randomization check is passed
```

2. Estimate the effect of advertising on sales in this campaign. (5 points)

a. Estimate the effectiveness of the ads by comparing all the subjects in the
treatment vs. control groups, regardless of whether they actually saw any ads
or not. This is known as the intent to treat (ITT) estimate.

b. How does your advertising effectiveness estimate change if you only
consider users (across both treatment and control groups) who actually saw
ads? This is known as the average treatment on the treated (ATT) estimate.

c. State your conclusions from the analysis in 2a and 2b.

```python
#2.a
ITT = treatment['sales'].mean() - control['sales'].mean()
Relative_ITT = (treatment['sales'].mean() - control['sales'].mean())/control['sales'].mean()
print(ITT, Relative_ITT)
```

```python
# 2.b
att_treatment_saw_ads = treatment[treatment['saw_ads'] == 1]
att_control_saw_ads = control[control['saw_ads'] == 1]

mean_sales_att_treat_filtered = att_treatment_saw_ads['sales'].mean()
mean_sales_att_control_filtered = att_control_saw_ads['sales'].mean()

actual_ATT = mean_sales_att_treat_filtered - mean_sales_att_control_filtered
actual_Relative_ATT = (mean_sales_att_treat_filtered - mean_sales_att_control_filtered) / mean_sales_att_control_filtered
print(actual_ATT, actual_Relative_ATT)

```

```python
#2.c
# from ITT analysis. being in the group where ads were shown led to an
# average increase in sales of approximately $0.0258 per user
# or about a 1.99% relative increase compared to the control group.
#from ATT analysis.
#By comparing the average sales of users who saw ads in the treatment group
#to those who saw ads in the control group
# we find a larger absolute increase of $0.0301 per user
# or a 1.94% relative increase.
#This effect is statistically significant
#meaning it's unlikely to have occurred by random chance.
# Both methods show positive impact
```

b. How do the ads affect non-Americans? Compute the ATT estimate (like in
2b) for people not living in America who saw ads.

c. State your conclusions from 3a and 3b. What does this imply for Uber Eats
and their ad budget allocation across different geographical areas?

```python
#3.a
att_treatment_saw_ads_am = treatment[(treatment['saw_ads'] == 1) & (treatment['country'] == 'american')]
att_control_saw_ads_am = control[(control['saw_ads'] == 1) & (control['country'] == 'american')]

att_tr_2 = att_treatment_saw_ads_am['sales'].mean()
att_ct_2 = att_control_saw_ads_am['sales'].mean()

att_new = att_tr_2 - att_ct_2
att_relative_new = (att_tr_2 - att_ct_2)/att_ct_2
print(att_new,att_relative_new)
```

```python
#3.b
att_treatment_saw_ads_am = treatment[(treatment['saw_ads'] == 1) & (treatment['country'] == 'nonamerican')]
att_control_saw_ads_am = control[(control['saw_ads'] == 1) & (control['country'] == 'nonamerican')]

att_tr_3 = att_treatment_saw_ads_am['sales'].mean()
att_ct_3 = att_control_saw_ads_am['sales'].mean()

att_nonam = att_tr_3 - att_ct_3
att_relative_nonam = (att_tr_3 - att_ct_3)/att_ct_3
print(att_nonam,att_relative_nonam)
```
